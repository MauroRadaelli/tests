```python
### Creating a numpy array from an image file

import imageio as im
image = im.imread('ferrari.jpg')
type(image)      
image.shape, image.dtype

print(image)
```

```python
### Displaying images - Increase contrast by setting min and max values

import matplotlib.pyplot as plt
plt.imshow(image, cmap=plt.cm.gray, vmin=30, vmax=200)        
# Remove axes and ticks
plt.axis('off')
```

```python
### Basic manipulations - Statistical information

image.mean(), image.max(), image.min()
```

```python
### Basic manipulations - Geometrical transformation

from scipy import ndimage

lx, ly, lz = image.shape

# Cropping
crop_image = image[lx // 4: - lx // 4, ly // 4: - ly // 4]

# up <-> down flip
flip_ud_image = np.flipud(image)

# rotation
rotate_image = ndimage.rotate(image, 45)
rotate_image_noreshape = ndimage.rotate(image, 45, reshape=False)


plt.imshow(crop_image)
plt.show()
plt.imshow(flip_ud_image)
plt.show()
plt.imshow(rotate_image)
plt.show()
plt.imshow(rotate_image_noreshape)
plt.show()
```

```python
### Image filtering - Blurring/smoothing

from scipy import ndimage
very_blurred = ndimage.gaussian_filter(image, sigma=7)

#import imageio as im
#im.imwrite('blurred_face.png', blurred_face) # uses the Image module (PIL)

import matplotlib.pyplot as plt
plt.imshow(very_blurred)
plt.show()
```

```python
### Image filtering - Sharpening

from scipy import ndimage
blurred_image = ndimage.gaussian_filter(image, 3)

filter_blurred_image = ndimage.gaussian_filter(blurred_image, 1)
alpha = 30
sharpened = blurred_image + alpha * (blurred_image - filter_blurred_image)

import matplotlib.pyplot as plt
plt.imshow(sharpened)
plt.show()
```
